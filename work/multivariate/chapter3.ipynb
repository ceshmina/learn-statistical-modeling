```python
import pandas as pd
```

```python
data = pd.read_csv('data/第03章/顧客満足度データ.csv', encoding='shift_jis')

data.columns = ['id', 'n_customers', 'cs_place', 'cs_equipment', 'cs_area', 'cs_trainer', 'n_trainers', 'is_training', 'is_benefit']

data.info()
data.head()
```

```python
import statsmodels.formula.api as smf
```

```python
res = smf.ols('n_customers ~ cs_place + cs_equipment + cs_area + cs_trainer', data).fit()
res.summary2()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# RのVIFはデフォルトで標準化するぽいので、ここの値は一致しない

data_ = data.drop(['id', 'n_customers'], axis=1)
n_cols = data_.shape[1]

vifs = [variance_inflation_factor(data_, i) for i in range(n_cols)]
vifs
```

```python
data2 = data[['cs_place', 'cs_equipment', 'cs_area', 'cs_trainer']]
n_cols = data2.shape[1]

vifs = [variance_inflation_factor(data2, i) for i in range(n_cols)]
vifs
```

```python
from sklearn.preprocessing import StandardScaler

cols = ['n_customers', 'cs_place', 'cs_equipment', 'cs_area', 'cs_trainer']
data3 = pd.DataFrame(StandardScaler().fit_transform(data[cols]), columns=cols)
data3.head()
```

```python
res3 = smf.ols('n_customers ~ cs_place + cs_equipment + cs_area + cs_trainer', data3).fit()
res3.summary2()
```

```python
data_ = data3.drop('n_customers', axis=1)
n_cols = data_.shape[1]

vifs = [variance_inflation_factor(data_, i) for i in range(n_cols)]
vifs
```

```python
res4 = smf.ols('n_customers ~ cs_place + cs_equipment + cs_area + cs_trainer + is_training + is_benefit', data).fit()
res4.summary2()
```
